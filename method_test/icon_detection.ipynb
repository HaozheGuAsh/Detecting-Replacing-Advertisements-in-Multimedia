### Cut center area and detect directly

```python
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import time


def icon_detect(threshHold=10, passLine=0.75, start=0, end=10, imshow=True, K=2, zoom=False, zoomProp=1/6):
    begin = time.clock()
    # Initialization
    width = 480
    height = 270
    frameOffset = 0
    frameLength = width * height * 3
    position = frameOffset * frameLength
    frameRate = 30

    videoPath = r'./data/data_test1.rgb'
    ## frameTemplatePath = [r'./data/mcdonalds_logo.jpg', r'./data/subway_snapshoot_2_zoomed.jpg']
    iconPathSet = [
        r'./data/subway_logo.jpg',
        r'./data/starbucks_logo.jpg'
        # r'./data/nfl_logo.jpg',
        # r'./data/mcdonalds_logo.jpg',
    ]
    iconSet = [ np.array(Image.open(iconPath)) for iconPath in iconPathSet ]
    iconInfoSet = []


    # Pre-process on brand-icons
    orb = cv2.ORB_create()
    for icon in iconSet:
        kpi, desi = orb.detectAndCompute(icon,None)
        iconInfoSet.append([kpi, desi])


    # Laod the .rgb video file into Memory
    file = open(videoPath, 'rb')    ## consider the opne-mode!

    # Detect brand-icons in video
    print("detecting...")
    ans = []
    frameOffset = start*30
    if zoom:
        frameSize = width * height
        pWidth = int(width * (1-2*zoomProp))
        pHeight = int(height * (1-2*zoomProp))
        framOffsetWidth = int(width*zoomProp)
        framOffsetHeight = int(height*zoomProp)
        frameArr = np.full((180,320,3), 0, np.uint8)
    else:
        frameArr = np.full((270,480,3), 0, np.uint8)
    while frameOffset <= end*30:
        # locate & laod frame
        file.seek(position, 0)
        frameByte = file.read(frameLength)

        frameOffset += 1
        position = frameOffset * frameLength

        # transfer .rgb frame
        if zoom:
            ind = framOffsetWidth + framOffsetHeight*width
            for y in range(pHeight):
                for x in range(pWidth):
                    frameArr[y][x] = [frameByte[ind], frameByte[ind + frameSize], frameByte[ind + frameSize*2]]
                    ind += 1
                ind += framOffsetWidth*2
        else:
            ind = 0
            for y in range(height):
                for x in range(width):
                    frameArr[y][x] = [frameByte[ind], frameByte[ind + width*height], frameByte[ind + width*height*2]]
                    ind += 1
        # get key-point info of current-frame
        kpf, desf = orb.detectAndCompute(frameArr,None)

        # compare with brand-icon info
        bf = cv2.BFMatcher()
        for info in iconInfoSet:
            matches = bf.knnMatch(info[1],desf, k=K)  ## what is 'k'?
            good = []
            for m,n in matches:
                if m.distance < passLine*n.distance:
                    good.append([m])
            if len(good) >= threshHold:
                ans.append( [iconInfoSet.index(info), frameOffset] )
                if imshow:
                    print("frameOffset =", frameOffset)
                    print("icon_ind =", iconInfoSet.index(info))
                    print("#good =", len(good))
                    ## print("max_similarity =", max([m.distance/n.distance for m,n in good]))
                    if zoom:
                        alignFrameArr = np.full((270,480,3), 0, np.uint8)
                        ## print("framOffsetWidth", framOffsetWidth, "pWidth+framOffsetWidth", pWidth+framOffsetWidth, "framOffsetHeight", framOffsetHeight, "pHeight+framOffsetHeight", pHeight+framOffsetHeight)
                        alignFrameArr[:pHeight, :pWidth] = frameArr
                        matchCompImg = cv2.drawMatchesKnn(iconSet[ iconInfoSet.index(info) ], info[0], alignFrameArr, kpf, good, None, flags=2)  ## what is 'flags'?
                    else:
                        matchCompImg = cv2.drawMatchesKnn(iconSet[ iconInfoSet.index(info) ], info[0], frameArr, kpf, good, None, flags=2)  ## what is 'flags'?
                    plt.figure("matchCompImg")
                    plt.imshow(matchCompImg)
                    plt.axis('off')
                    plt.show()
    print("Detection finished in {} seconds.".format(time.clock()-begin))
    return ans

## icon_detect(threshHold=12, passLine=0.75, start=0, end=75, imshow=True)
## icon_detect(threshHold=12, passLine=0.75, start=65, end=75, imshow=True, zoom=True)
## icon_detect(threshHold=10, passLine=0.75, start=170, end=185, imshow=True)
icon_detect(threshHold=10, passLine=0.75, start=170, end=185, imshow=True, zoom=True)
```
